```python
import os

app_home = os.environ.get('APP_HOME')
```

```python
import boto3

sm_client = boto3.client(
    'secretsmanager',
    region_name='us-east-1'
)

secret_token = sm_client.get_secret_value(SecretId='gmail_token')['SecretBinary']
```

```python
import pickle

creds = pickle.loads(secret_token)

from googleapiclient.discovery import build

service = build('gmail', 'v1', credentials=creds)
```

```python
START_DATE = '2022/01/01'
END_DATE = '2022/01/02'
```

```python
messages_all = []
next_page_token = None

users = service.users()
while True:
    print(f'Processing in range between {START_DATE} and {END_DATE} using token {next_page_token}')
    if next_page_token:
        messages = users. \
            messages(). \
            list(
                userId='me', 
                q=f'after:{START_DATE} before:{END_DATE}',
                pageToken=next_page_token
            ). \
            execute()
        messages_all += messages['messages']
        next_page_token = messages.get('nextPageToken')
    else:
        messages = users. \
            messages(). \
            list(
                userId='me', 
                q=f'after:{START_DATE} before:{END_DATE}'
            ). \
            execute()
        messages_all = messages['messages']
        next_page_token = messages.get('nextPageToken')
    if next_page_token == None:
        break
        
```

```python
message_list = []

for m in messages_all:
    msg = users.messages().get(userId='me', id=m['id']).execute()
    message_list.append(msg)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(message_list)
```
